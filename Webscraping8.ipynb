Create a Python program that scrapes the information from the NYKAA App for face serum offers ranging from 40% And Above.

a) Name of the brand
 
b) Price

c)Discount 

d) Offer Price


```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
page = requests.get("https://www.nykaa.com/search/result/?q=face%20serum%20&root=search&sourcepage=home&searchType=history&suggestionType=query&ssp=1&searchItem=face%20serum%20&discount_range_filter=40-*")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

name of the brand 

```python
Name = []
for i in soup.find_all('div',class_="css-xrzmfa"):
    Name.append(i.text)
Name
```

original price 


```python
Price = []
for i in soup.find_all('span',class_="css-17x46n5"):
    Price.append(i.text)
Price
```

discount


```python
Discount = []
for i in  soup.find_all('span',class_="css-cjd9an"):
    Discount.append(i.text)
Discount
```

 
offer price 


```python
Offer_Price = []
for i in  soup.find_all('span',class_="css-111z9ua"):
    Offer_Price.append(i.text)
Offer_Price
```

```python
print(len(Name),len(Price),len(Discount),len(Offer_Price))
```

```python
df = pd.DataFrame({'BRAND NAME':Name,'PRICE':Price,"DISCOUNT":Discount,'OFFER PRICE':Offer_Price}
            )
df
```
