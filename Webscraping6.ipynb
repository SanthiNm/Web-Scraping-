```python
Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.
b) Top 10 women’s ODI Batting players along with the records of their team and rating.
c) Top 10 women’s ODI all-rounder along with the records of their team and rating.
```

```python
from bs4 import BeautifulSoup
import pandas as pd 
import requests
```

```python
page = requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
#team 
team = []
for i in soup.find_all('span',class_="u-hide-phablet"):
    team.append(i.text)
Team = team[:10]
Team
```

```python
#matches 
Values = []
for j in soup.find_all('td',class_="table-body__cell u-center-text"):
       Values.append(j.text)

Values
```

```python
Matches = ['24']
Point = ['4290']
for i in range(1,len(Values)-1,2):
    Matches.append(Values[i+1])
    Point.append(Values[i])
```

```python

Match = Matches[:10]
Match
```

```python
Points = Point[:10]
Points
```

```python
#RATING
rating = ['116']
for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
     rating.append(i.text)
Rating = rating[:10]   
Rating
```

```python
print(len(Team),len(Match),len(Points),len(Rating))
```

```python
df = pd.DataFrame({"TEAM":Team,"MATCHES":Match,"POINTS":Points,"RATING":Rating})
df
```

```python
Batsman = batsman[:10]
Batsman
```

```python
team = ['ENG']
for i in soup1.find_all('span',class_="table-body__logo-text"):
     team.append(i.text)
```

```python
Team = team[:10]
Team
```

```python
rating = ['807']
for i in soup1.find_all('td',class_="table-body__cell rating"):
    rating.append(i.text)
rating
```

```python
Rating = rating[:10]
Rating
```

```python
print(len(Batsman),len(Team),len(Rating))
```

```python
df=pd.DataFrame({"BATSMAN":Batsman,"TEAM":Team,"RATING":Rating})
df
```
